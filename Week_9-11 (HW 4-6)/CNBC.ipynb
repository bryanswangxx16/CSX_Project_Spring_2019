```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
import numpy as np
import requests
import time
from bs4 import BeautifulSoup
import pandas_datareader.data as web
import re
```

```python
url_links = []

sections = ['business', 'trading-nation', 'earnings', 'economy', 'finance', 'health-and-science', 'real-estate', 'energy', 'transportation', 'industrials', 'retail', 'wealth', 'small-business']

for sector in sections:
    print(sector)
    for i in range(1, 10):
        url = 'https://www.cnbc.com/' + sector + '/?page=' + str(i)
        html = requests.get(url)
        content = html.content
    #     content.decode().strip().split(';')
        soup = BeautifulSoup(content, "html.parser")
        for tags in soup.find_all('a'):
            try:
                if re.search('html', tags['href']):
                    url_links.append(tags['href'])
            except:
                continue

# some linkes may be duplicated thus we only select those that only appear once
final_urls = []
for url in url_links:
    if url not in final_urls:
        final_urls.append(url)
final_urls[:5]
```

```python
len(final_urls)
```

```python
title_all = []
time_all = []
content_all = []
url_all = []

for url in final_urls:
    try:
        link = url
        page = requests.get(link).content
        soup = BeautifulSoup(page, "html.parser")
        newsTitle = soup.title.text
        print(newsTitle.lstrip())
        print(link)
        newsTime = soup.find_all("time", {"data-testid": 'published-timestamp'})[0].text


        newsContent = ''
        soup = soup.find_all("div", {"class":'group'})
        for seg in soup:
            for tag in seg.find_all('p'): 
                newsContent += tag.text

        title_all.append(newsTitle)
        time_all.append(newsTime)
        content_all.append(newsContent)
        url_all.append(link)
        if len(title_all) % 100 == 0:
            print()
            print("{:=^101}".format(len(title_all))) 
            print()
        if len(title_all) % 20 == 0:
            file = pd.DataFrame({'Title' : title_all, 'Time':time_all, 'Content':content_all, 'Link':url_all})
            file['Date'] = [x.split(',')[-1].split('  ')[0] for x in file['Time']]
            file['Date'] = pd.to_datetime(file['Date'])
            file.to_csv('CNBC.csv')
    except:
        continue

file.head()
```
