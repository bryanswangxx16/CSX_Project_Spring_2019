```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
import numpy as np
import requests
import time
from bs4 import BeautifulSoup
import pandas_datareader.data as web
import re
```

```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup as bs
from selenium.webdriver.common.by import By

# I used Firefox; you can use whichever browser you like.
driver = webdriver.Chrome('/mnt/c/Users/bryan/OneDrive/Desktop/chromedriver/chromedriver.exe')


final_urls = []

sections = ['markets', 'banking-insurance', 'fintech', 'etfs-mutual-funds', ]

for sector in sections:
    driver.get('https://www.forbes.com/' + sector)
    print(sector, end = ' ')
    for i in range(10):
        print(i, end = ' ')
        try:
            more_buttons = driver.find_element_by_class_name("load-more-bg")
            more_buttons.click()
            time.sleep(1)
        except:
            continue
    time.sleep(3)
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, "html.parser")
    url_links = []
    for tags in soup.find_all('a'):
        try:
            if re.search('https://www.forbes.com/sites', tags['href']):
                if len(tags['href'].split('/')) >= 8:
                    url_links.append(tags['href'])
        except:
            continue

    # some linkes may be duplicated thus we only select those that only appear once
    for url in url_links:
        if url not in final_urls:
            final_urls.append(url)
final_urls
```

```python
title_all = []
time_all = []
content_all = []
url_all = []

for url in final_urls:
    try:
        link = url
        page = requests.get(link).content
        soup = BeautifulSoup(page, "html.parser")
        newsTitle = soup.title.text
        print(newsTitle.lstrip())
        print(link)
        newsTime = soup.find_all('meta', {'property':"og:updated_time"})[0]['content']

        newsContent = ''
        soup = soup.find_all('div', {'class':'yiv2271847414WordSection1'})
        for seg in soup:
            for tag in seg.find_all('span', {'style':"font-family: arial, helvetica"}): 
                newsContent += tag.text

        title_all.append(newsTitle)
        time_all.append(newsTime)
        content_all.append(newsContent)
        url_all.append(link)
        if len(title_all) % 20 == 0:
            print()
            print("{:=^101}".format(len(title_all))) 
            print()
        if len(title_all) % 20 == 0:
            file = pd.DataFrame({'Title' : title_all, 'Time':time_all, 'Content':content_all, 'Link':url_all})
            file['Date'] = [x.split(',')[-1].split('  ')[0] for x in file['Time']]
            file['Date'] = pd.to_datetime(file['Date'])
            file.to_csv('Forbes.csv')
    except:
        continue
        
        
file = pd.DataFrame({'Title' : title_all, 'Time':time_all, 'Content':content_all, 'Link':url_all})
file['Date'] = [x.split(',')[-1].split('  ')[0] for x in file['Time']]
file['Date'] = pd.to_datetime(file['Date'])
file.to_csv('Forbes.csv')
file.head()
```
